```python
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt

fr=np.array([50,80,110,140,170,200,230,260,300])
a=np.array([1.75,1.8,2.2,2.4,2.1,1.75,1.25,0.9,0.65])

X_Y_Spline = make_interp_spline(fr, a)
X_ = np.linspace(fr.min(), fr.max(), 500)
Y_ = X_Y_Spline(X_)

plt.plot(X_,Y_)

plt.ylim(ymin=0)
plt.xlim(xmin=50,xmax=300)
plt.style.use("seaborn")
plt.scatter(fr,a,color='red',linewidths=0.2,alpha=0.5)
plt.xlabel("frequence moteur (Hz)")
plt.ylabel("amplitude (cm)")
plt.savefig('forte.png')
```
